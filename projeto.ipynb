```python
import pandas as pd
import pandera as pa
```

### Extração dos dados (E)
- sep -> indica que o separador é um ';'
- parse_dates=['variavel'] -> indica a(s) que precisa(m) ser convertida(s) para o tipo data
- "dayfirst=True" indica que o dia vem antes do mês

```python
df = pd.read_csv('./ocorrencia_2010_2020.csv', sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = { # Valida a coluna "codigo_ocorrencia" do DataFrame
        "codigo":pa.Column(pa.Int, required=False),
        "codigo_ocorreencia":pa.Column(pa.Int), # valida se a coluna indicada possui dados do tipo inteiro
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificação":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2)),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True), # dentro do Check, entra uma expressão 'regular' => pesquisar, pois é inportante para valiadação
    }
)

schema.validatet(df)
```
