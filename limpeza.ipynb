```python
import pandas as pd
```

```python
df = pd.read_csv('./ocorrencia_2010_2020.csv', sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
df.head()
```

### **Manipulando os dados**

```python
df.loc[1,'ocorrencia_cidade']
```

```python
df.loc[1:3]
```

```python
df.loc[[10,40]] # indica que é somente a linha
```

```python
df.loc[:,'ocorrencia_cidade'] # trás todos os dados de uma coluna sem a ncessidade de especificar as linhas

# loc[:{os ':' trás todos os dados de uma coluna},{nome ou indice da coluna}]
```

> indendifica se a coluna 'codigo_ocorrencia' é unica, ou seja, os valores não repetem

```python
df.codigo_ocorrencia.is_unique
```
