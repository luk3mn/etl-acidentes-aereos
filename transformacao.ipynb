```python
import pandas as pd
import pandera as pa
```

### **Extração dos dados (E)**
- "na_values" => ja faz a atribuição do NaN aos dados inconsistentes leitura dos dados

```python
valores_ausentes = ['**','###!','####','****','*****','NULL']
df = pd.read_csv('./ocorrencia_2010_2020.csv', sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

> Validação

```python
schema = pa.DataFrameSchema(
    columns = { # Valida a coluna "codigo_ocorrencia" do DataFrame
        "codigo_ocorrencia":pa.Column(pa.Int), # valida se a coluna indicada possui dados do tipo inteiro
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo":pa.Column(pa.String, nullable=True),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True), # dentro do Check, entra uma expressão 'regular' => pesquisar, pois é inportante para valiadação
        "total_recomendacoes":pa.Column(pa.Int)
    }
)

schema.validate(df)
```

> Combinando Filtros

```python
# ocorrências cuja a classificação == INCIDENTE GRAVE e o estado == SP
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
# ocorrências cuja a (classificação == INCIDENTE GRAVE ou classificação == INCIDENTE) e estado == SP
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

- segunda forma de filtrar usando o OR com o método isin e passando a lista de atributos

```python
# ocorrências cuja a (classificação == INCIDENTE GRAVE ou classificação == INCIDENTE) e estado == SP
filtro1 = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

> Aplicando outras formas de filtros

```python
# ocorrências cuja cidade comecem com a letra C
filtro = df.ocorrencia_cidade.str[0] == 'C'
df.loc[filtro]
```

```python
# ocorrência cjua cidade terminam com a letra A
filtro = df.ocorrencia_cidade.str[-1]  == 'A'
df.loc[filtro]
```

```python
# ocorrência cuja cidade terminam com os caracteres MA
filtro = df.ocorrencia_cidade.str[-2:] == 'MA'
df.loc[filtro]
```

```python
# ocorrência cuja cidade contêm )em qualquer parte do conteúdo) os caracteres MA
filtro = df.ocorrencia_cidade.str.contains('MA')
df.loc[filtro]
```

```python
# ocorrência cuja cidade contêm )em qualquer parte do conteúdo) os caracteres MA e Al
filtro = df.ocorrencia_cidade.str.contains('MA|AL')
df.loc[filtro]
```

```python
# ocorrência do ano de 2015
filtro = df.ocorrencia_dia.dt.year == 2015
df.loc[filtro]
```

```python
# ocorrências do ano de 2015 e mês 12
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
df.loc[filtro1 & filtro2]
```

- aplicando o filtro em uma linha

```python
# ocorrências do ano de 2015 e mês 12
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12)
df.loc[filtro]
```

```python
# ocorrências do ano de 2015 e mês 12 e dia 08
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
filtro3 = df.ocorrencia_dia.dt.day == 8
df.loc[filtro1 & filtro2 & filtro3]
```
